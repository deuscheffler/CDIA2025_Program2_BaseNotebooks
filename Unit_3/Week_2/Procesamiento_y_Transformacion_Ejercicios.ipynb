```python
def minmax(col):
    return round(col - col.min()) / (col.max() - col.min(), 4)
```

```python
def zscore(col):
    return round((col - col.mean()) / col.std(), 4)
```

```python
display(data_iris.head())

iris_norm= data_iris.copy()
iris_norm=iris_norm.apply(minmax, axis=0)
```
